# 查詢
在 BST 中查詢是否有 target 這個值，有則回傳此值，沒有則回傳 None。

```python
def search(self,root,target): # root為數列第一個數值 target為要查詢之值
    if target < root.val: # 若查詢之值小於root之值
        if root.left == None: # 往左找此數 若左邊沒有數 則樹中無此查詢之值
            return None # 回傳None
        else: # 往左找此數 若左邊有數
            p = root.left # 則以root左邊為p
            return self.search(p,target) # 以p作為search函式的root重新跑函式
    elif target > root.val: # 若查詢之值大於root之值
        if root.right == None: # 往右找此數 若右邊沒有數 則樹中無此查詢之值
            return None # 回傳None
        else: # 往右找此數 若右邊有數
            p = root.right # 則以root右邊為p
            return self.search(p,target) # 以p作為search函式的root重新跑函式
    else: #若查詢之值等於root之值
        return root # 回傳查詢之值
```

# 新增
在 BST 中新增 val 這個值，新增之值一定在 BST 的尾端（不會有孩子）。

```python
def insert(self,root,val): # root為數列第一個數值 val為要新增之值
    if val <= root.val: # 若新增之值小於等於root之值
        if root.left == None: # 往左找此數 若左邊沒有數 則放val在root左邊
            root.left = TreeNode(val)
        else: # 若左邊有數
            p = root.left # 則以root左邊為p
            self.insert(p,val) # 以p作為insert函式的root重新跑函式
    else: # 若新增之值大於root之值
        if root.right == None: # 往右找此數 若右邊沒有數 則放val在root右邊
            root.right = TreeNode(val)
        else: # 若右邊有數
            p = root.right # 則以root右邊為p
            self.insert(p,val) # 以p作為insert函式的root重新跑函式
```

# 刪除
在 BST 中尋找有無 target 這個值，若無則 pass，若有則將此值刪掉，若此值有兩個以上，則重複刪掉。

```python
def delete(self,root,target): # root為數列第一個數值 target為要刪除之值
    i = root # i等於root 為數列第一個數值
    parent = root
    #找parent、x
    while target != root.val: #target與root之值不相等時
        parent = root
        if target < root.val: 
            if root.left == None: #若樹中無此值 則pass
                break
            else:
                root = root.left # 以root左邊為root 回去跑while迴圈
        elif target > root.val:
            if root.right == None: #若樹中無此值 則pass
                break
            else:
                root = root.rightt # 以root右邊為root 回去跑while迴圈
    #跑完while後 若樹中有此值 root會成為target的位置 parent會成為target的父節點
                    
    if target == root.val: # 樹中有此值
        if root.left == None and root.right == None: #當target沒有孩子
            if parent.val < root.val: # 若target的父節點小於target
                parent.right = None # target的父節點指向None
            elif parent.val > root.val:
                parent.left = None
            else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                del root.val #刪除值
        elif root.left == None and root.right != None: #target沒有左孩子 有右孩子
            if parent.val < root.val: 
                parent.right = root.right # target的父節點指向target的右孩子
            elif parent.val > root.val:
                parent.left = root.right # target的父節點指向target的右孩子
            else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                parent = root
                root = root.right
                if root.left == None and root.right == None: #target.right沒有孩子
                    parent.val = root.val
                    parent.right = None
                elif root.left == None and root.right != None: #target.right沒有左孩子 有右孩子
                    parent.val = root.val
                    parent.right = root.right
                elif root.left != None and root.right == None: #target.right有左孩子 沒有右孩子
                    temp = root
                    while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                        Minparent = temp
                        temp = Minparent.left

                    Min = temp
                    parent.val = Min.val #將欲刪除之值換成最小值
                    if Min.right != None: 
                            Minparent.left = Min.right
                    else:
                        Minparent.left = None
                else: #target.right有兩個孩子
                    temp = root
                    while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                        Minparent = temp
                        temp = Minparent.left

                    Min = temp
                    parent.val = Min.val
                    if Min.right != None: #將欲刪除之值換成最小值
                        Minparent.left = Min.right
                    else:
                        Minparent.left = None
                        
                        
        elif root.left != None and root.right == None: #target有左孩子 沒有右孩子
            if parent.val < root.val:
                parent.right = root.left # target的父節點指向target的左孩子
            elif parent.val > root.val:
                parent.left = root.left # target的父節點指向target的左孩子
            else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                parent = root
                root = root.left
                if root.left == None and root.right == None: #target.left沒有孩子
                    parent.val = root.val
                    parent.left = None
                elif root.left == None and root.right != None: #target.left沒有左孩子 有右孩子
                    temp = root
                    while temp.right != None: #找target左邊的最大值 以及最大值的父節點
                        Maxparent = temp
                        temp = Maxparent.right

                    Max = temp
                    parent.val = Max.val #將欲刪除之值換成最大值
                    if Max.left != None:
                        Minparent.right = Max.left
                    else:
                        Maxparent.right = None
                elif root.left != None and root.right == None: #target.left有左孩子 沒有右孩子
                    parent.val = root.val
                    parent.left = root.left
                else: #target.left有兩個孩子
                    temp = root
                    while temp.right != None: #找target左邊的最大值 以及最大值的父節點
                        Maxparent = temp
                        temp = Maxparent.right

                    Max = temp
                    parent.val = Max.val #將欲刪除之值換成最大值
                    if Max.left != None:
                        Maxparent.right = Max.left
                    else:
                        Maxparent.right = None 

        else: #target有兩個孩子
            temp = root.right
            if temp.left == None:
                Minparent = root

                Min = temp
                root.val = Min.val
                if Min.right != None:
                    Minparent.right = Min.right
                else:
                    Minparent.right = None
            else:
                while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                    Minparent = temp
                    temp = Minparent.left

                Min = temp
                root.val = Min.val #將欲刪除之值換成最小值
                if Min.right != None:
                    Minparent.left = Min.right 
                else:
                    Minparent.left = None
        self.delete(i,target) #呼叫delete函式 以檢查有無重複值還未刪除掉
    return i #回傳修改過後的樹的第一個父節點
```

# 修改
在 BST 中尋找有無 target 這個值，若無則 pass，若有則將此值刪掉，並新噌 new_val，刪掉了幾個重複的 target 就會新增幾個重複的 new_val。

```python
def modify(self,root,target,new_val):
    #先刪除
    i = root
    #當在樹中找得到target時 跑迴圈
    while self.search(root,target) != None:
        parent = root
        #找parent、x
        while target != root.val: #target與root之值不相等時
            parent = root
            if target < root.val: 
                if root.left == None: #若樹中無此值 則pass
                    break
                else:
                    root = root.left # 以root左邊為root 回去跑while迴圈
            elif target > root.val:
                if root.right == None: #若樹中無此值 則pass
                    break
                else:
                    root = root.rightt # 以root右邊為root 回去跑while迴圈
        #跑完while後 若樹中有此值 root會成為target的位置 parent會成為target的父節點

        if target == root.val: # 樹中有此值
            if root.left == None and root.right == None: #當target沒有孩子
                if parent.val < root.val: # 若target的父節點小於target
                    parent.right = None # target的父節點指向None
                elif parent.val > root.val:
                    parent.left = None
                else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                    del root.val #刪除值
            elif root.left == None and root.right != None: #target沒有左孩子 有右孩子
                if parent.val < root.val: 
                    parent.right = root.right # target的父節點指向target的右孩子
                elif parent.val > root.val:
                    parent.left = root.right # target的父節點指向target的右孩子
                else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                    parent = root
                    root = root.right
                    if root.left == None and root.right == None: #target.right沒有孩子
                        parent.val = root.val
                        parent.right = None
                    elif root.left == None and root.right != None: #target.right沒有左孩子 有右孩子
                        parent.val = root.val
                        parent.right = root.right
                    elif root.left != None and root.right == None: #target.right有左孩子 沒有右孩子
                        temp = root
                        while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                            Minparent = temp
                            temp = Minparent.left

                        Min = temp
                        parent.val = Min.val #將欲刪除之值換成最小值
                        if Min.right != None: 
                                Minparent.left = Min.right
                        else:
                            Minparent.left = None
                    else: #target.right有兩個孩子
                        temp = root
                        while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                            Minparent = temp
                            temp = Minparent.left

                        Min = temp
                        parent.val = Min.val
                        if Min.right != None: #將欲刪除之值換成最小值
                            Minparent.left = Min.right
                        else:
                            Minparent.left = None


            elif root.left != None and root.right == None: #target有左孩子 沒有右孩子
                if parent.val < root.val:
                    parent.right = root.left # target的父節點指向target的左孩子
                elif parent.val > root.val:
                    parent.left = root.left # target的父節點指向target的左孩子
                else: # 這裡的parent為原本的root 也就是第一個值 這行的意思為若target為樹中第一個數
                    parent = root
                    root = root.left
                    if root.left == None and root.right == None: #target.left沒有孩子
                        parent.val = root.val
                        parent.left = None
                    elif root.left == None and root.right != None: #target.left沒有左孩子 有右孩子
                        temp = root
                        while temp.right != None: #找target左邊的最大值 以及最大值的父節點
                            Maxparent = temp
                            temp = Maxparent.right

                        Max = temp
                        parent.val = Max.val #將欲刪除之值換成最大值
                        if Max.left != None:
                            Minparent.right = Max.left
                        else:
                            Maxparent.right = None
                    elif root.left != None and root.right == None: #target.left有左孩子 沒有右孩子
                        parent.val = root.val
                        parent.left = root.left
                    else: #target.left有兩個孩子
                        temp = root
                        while temp.right != None: #找target左邊的最大值 以及最大值的父節點
                            Maxparent = temp
                            temp = Maxparent.right

                        Max = temp
                        parent.val = Max.val #將欲刪除之值換成最大值
                        if Max.left != None:
                            Maxparent.right = Max.left
                        else:
                            Maxparent.right = None 

            else: #target有兩個孩子
                temp = root.right
                if temp.left == None:
                    Minparent = root

                    Min = temp
                    root.val = Min.val
                    if Min.right != None:
                        Minparent.right = Min.right
                    else:
                        Minparent.right = None
                else:
                    while temp.left != None: #找target右邊的最小值 以及最小值的父節點
                        Minparent = temp
                        temp = Minparent.left

                    Min = temp
                    root.val = Min.val #將欲刪除之值換成最小值
                    if Min.right != None:
                        Minparent.left = Min.right 
                    else:
                        Minparent.left = None

        self.insert(i,new_val) #呼叫insert函式以新增new_value

    return i #回傳修改過後的樹的第一個父節點
```
